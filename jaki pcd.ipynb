## import library

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## deteksi tepi

```python
image = cv.imread('jaki.jpeg')
```

```python
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray, 75, 150)

plt.figure(figsize=(5,5))
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
contours, hierarchy = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

contour_img = image_rgb.copy()

cv.drawContours(contour_img, contours, -1, (0,255,0), 2)

plt.figure(figsize=(5,5))
plt.imshow(contour_img)
plt.title("Contours Detection")
plt.axis("off")
plt.show()
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))

axs[0].imshow(image_rgb)
axs[0].set_title("Original Image")

axs[1].imshow(edges, cmap="gray")
axs[1].set_title("Canny Edge Detection")

axs[2].imshow(contour_img)
axs[2].set_title("Contours Detection")

for a in axs:
    a.axis("off")

plt.tight_layout()
plt.show()
```
